```python
import pandas as pd

df = pd.read_csv('telecom_X_limpio.csv')

df.head()
```

# 2. Eliminación de Columnas Irrelevantes

```python
columnas_a_eliminar = ['ID_Cliente', 'customerID', 'Cuentas_Diarias']
```

```python
df = df.drop(columns=columnas_a_eliminar, errors='ignore')

df.head()
```

# 3. Encoding

```python
df_final = pd.get_dummies(df, columns=['Tipo_Contrato', 'Metodo_Pago', 'Plan_Internet'], dtype=int)
```

```python
print("Nuevas columnas después del Encoding:")
print(df_final.columns)

df_final.head()
```

```python
df_final.info()
```

#4. Verificación de la Proporción de Cancelación (Churn)

```python
proporcion_churn = df_final['Churn'].value_counts(normalize=True) * 100
```

```python
conteo_churn = df_final['Churn'].value_counts()
```

```python

print("--- Análisis de Desbalance ---")
print(f"Clientes que se QUEDAN (0): {conteo_churn[0]} ({proporcion_churn[0]:.2f}%)")
print(f"Clientes que se VAN    (1): {conteo_churn[1]} ({proporcion_churn[1]:.2f}%)")
```

HALLAZGO:

La clase mayoritaria (clientes que permanecen) representa aproximadamente el 73% de los datos, mientras que la clase minoritaria (clientes que cancelan) representa el 27%.

Estado del Dataset:

Existe un desbalance moderado. Aunque no es un desbalance extremo, es suficiente para que debamos prestar atención a las métricas de evaluación más adelante.

# 5. Balanceo de Clases (opcional)

```python
X = df_final.drop('Churn', axis=1)
y = df_final['Churn']

smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)

print("--- Datos después del balanceo (SMOTE) ---")
print(y_resampled.value_counts(normalize=True) * 100)
```

```python
df_balanceado = pd.DataFrame(X_resampled, columns=X.columns)
df_balanceado['Churn'] = y_resampled


correlaciones = df_balanceado.corr()['Churn'].sort_values(ascending=False)

print("Variables que REALMENTE influyen en la cancelación:")
print(correlaciones[1:11].round(2))
```

#6. Normalización o Estandarización (si es necesario)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X_resampled)

X_final = pd.DataFrame(X_scaled, columns=X.columns)

print("¡Listo! Datos normalizados y listos para los modelos.")
X_final[['Cargo_Mensual', 'Meses_Cliente']].head().round(2)
```

# 7. Análisis de Correlación

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

sns.heatmap(df_balanceado.corr(), annot=False, cmap='coolwarm', linewidths=0.5)

plt.title('Matriz de Correlación - Identificación de Variables Clave para Churn')
plt.show()
```

#8. Análisis Dirigido

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.boxplot(x='Churn', y='Meses_Cliente', data=df_final, ax=axes[0], palette='magma')
axes[0].set_title('Distribución de Meses del Cliente por Cancelación')
axes[0].set_xticklabels(['Se queda (0)', 'Se va (1)'])

sns.kdeplot(df_final[df_final['Churn'] == 0]['Cargo_Total'], label='Se queda', shade=True, ax=axes[1])
sns.kdeplot(df_final[df_final['Churn'] == 1]['Cargo_Total'], label='Se va', shade=True, ax=axes[1])
axes[1].set_title('Densidad de Gasto Total por Cancelación')
axes[1].legend()

plt.tight_layout()
plt.show()
```

#9. Separación de Datos

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_final, y_resampled, test_size=0.20, random_state=42)

print(f"--- División Completada ---")
print(f"Muestras para Entrenamiento: {X_train.shape[0]}")
print(f"Muestras para Prueba: {X_test.shape[0]}")
```

# 10. Creación de Modelos

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

modelo_logistico = LogisticRegression(random_state=42)
modelo_logistico.fit(X_train, y_train)

modelo_forest = RandomForestClassifier(random_state=42, n_estimators=100)
modelo_forest.fit(X_train, y_train)

print("--- Entrenamiento Finalizado ---")
print("Modelo 1: Regresión Logística entrenado.")
print("Modelo 2: Random Forest entrenado.")
```

JUSTIFICACIÓN

Regresión Logística: Este modelo es como una balanza muy sensible. Si le pones una variable con números grandes (como el Gasto Total que son miles) y otra con números pequeños (como 0 o 1), la balanza se inclina por error hacia el número más grande. Al normalizar, hacemos que todos los datos pesen lo mismo para que la balanza sea justa.

Random Forest: Este modelo funciona haciendo preguntas de "Sí" o "No" (como un árbol de decisiones). A él no le importa si el número es 1 o 1,000,000; solo le interesa si es mayor o menor a cierto punto. Por eso, no necesita normalización, aunque no le hace daño si ya la tiene.

#11. Evaluación de los Modelos

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

modelos = {'Regresión Logística': modelo_logistico, 'Random Forest': modelo_forest}

for nombre, modelo in modelos.items():
    predicciones = modelo.predict(X_test)

    print(f"\n--- Evaluación: {nombre} ---")
    print(f"Exactitud (Accuracy): {accuracy_score(y_test, predicciones):.2f}")
    print(f"Precisión:           {precision_score(y_test, predicciones):.2f}")
    print(f"Recall:              {recall_score(y_test, predicciones):.2f}")
    print(f"F1-Score:            {f1_score(y_test, predicciones):.2f}")

    print("Matriz de Confusión:")
    print(confusion_matrix(y_test, predicciones))
```

ANÁLISIS DE RESULTADOS Y COMPARACIÓN

1. Desempeño General:

Ambos modelos tuvieron un desempeño excelente, con una exactitud (Accuracy) muy similar: 84% para la Regresión Logística y 85% para el Random Forest. Esto significa que aciertan en casi 9 de cada 10 casos.

2. ¿Cuál es el mejor modelo?

El ganador por muy poco es el Random Forest. No solo tiene un 1% más de exactitud, sino que su Recall (0.86) es superior al de la Regresión Logística (0.84).

¿Por qué importa el Recall? Porque el Recall mide qué tan bueno es el modelo para encontrar a los clientes que realmente se van. En este negocio, es mejor tener una "falsa alarma" que dejar que un cliente se escape sin darnos cuenta.

3. Sobre el Overfitting y Underfitting:

No hay Underfitting: Los modelos aprendieron bien los patrones, ya que las métricas son altas y no se quedan en valores mediocres (como un 50% o 60%).

No parece haber Overfitting grave: Como los resultados en este conjunto de "Prueba" son consistentes y altos (0.85), el modelo demuestra que puede generalizar bien con datos nuevos que nunca había visto, y no solo memorizó los datos de entrenamiento.

4. Conclusión:

Elegiría el Random Forest para implementar en la empresa, ya que es el más equilibrado y el que mejor detecta los casos positivos de cancelación (Churn).

Conclusión del análisis:

INFORME DE ANÁLISIS DE CANCELACIÓN DE CLIENTES (CHURN) - TELECOM X

1. Introducción

El objetivo de este proyecto fue identificar los motivos por los cuales los clientes abandonan la compañía y construir un modelo de Inteligencia Artificial capaz de predecir estas cancelaciones antes de que ocurran. Tras limpiar los datos, balancear las clases y entrenar modelos, estos son los hallazgos principales.

2. Factores Clave que Influyen en la Cancelación

Basado en la Matriz de Correlación y la Importancia de Variables de nuestros modelos, los tres factores que más "empujan" a un cliente a irse son:

Tipo de Contrato: Los clientes con contratos mes a mes tienen una probabilidad de cancelación muchísimo más alta que aquellos con contratos a uno o dos años.

Método de Pago e Internet: El uso de Fibra Óptica (curiosamente asociado a más quejas o costos) y el pago mediante Cheque Electrónico aparecen como indicadores críticos de abandono.

Antigüedad (Tenure): Existe una relación inversa clara; los clientes que llevan menos de 6 meses en la empresa son los más propensos a irse. Una vez que el cliente supera los 20 meses, la lealtad aumenta considerablemente.

3. Rendimiento de los Modelos

Para este análisis comparamos dos modelos: Regresión Logística y Random Forest.

Modelo Ganador: El Random Forest obtuvo el mejor desempeño con una exactitud del 85%.

Capacidad de Detección: Lo más importante es su Recall de 0.86, lo que significa que el modelo es capaz de detectar correctamente al 86% de los clientes que tienen intención de cancelar. Esto nos permite actuar a tiempo en la gran mayoría de los casos.

Generalización: El modelo no presenta overfitting, ya que sus resultados en los datos de prueba son consistentes con los de entrenamiento.

4. Estrategias de Retención Propuestas

Basándome en los datos obtenidos, propongo las siguientes acciones comerciales:

Plan de Migración de Contratos: Crear promociones u ofertas exclusivas para que los clientes de "mes a mes" se cambien a contratos de 1 año. Esto aumenta la "barrera de salida" y estabiliza la base de clientes.

Programa de Bienvenida (Primeros 6 meses): Dado que los clientes nuevos se van más rápido, se debe implementar un seguimiento intensivo durante el primer semestre, ofreciendo soporte prioritario o descuentos por lealtad temprana.

Revisión del Servicio de Fibra Óptica: Investigar por qué los usuarios de fibra óptica están más insatisfechos que los de ADSL. Podría haber un problema de precio o de estabilidad técnica en esa tecnología específica.

Incentivos para Pagos Automáticos: Fomentar el cambio de "Cheque Electrónico" a "Tarjeta de Crédito o Débito Automática", ya que los pagos automáticos reducen la fricción y están asociados a una menor tasa de cancelación.
